```python
#imports 
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pyarrow.parquet as pq
import glob
import os
import math
```

```python
# Folder containing videos
video_dir = "/home/projects/sipl-prj10496/project_files/data/source_videos/NL129" 
extensions = ("*.mp4", "*.avi", "*.mov", "*.mkv")

video_paths = []
for pattern in extensions:
    video_paths.extend(glob.glob(os.path.join(video_dir, pattern)))

print(f"Found {len(video_paths)} videos")

for video_path in sorted(video_paths):
    print(f"Reading: {video_path}")
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"  Could not open: {video_path}")
        continue

    # Read frames here (example reads the first frame)
    ret, frame = cap.read()
    if not ret:
        print(f"  No frames in: {video_path}")
    else:
        # Example: get frame size
        h, w = frame.shape[:2]
        print(f"  First frame size: {w}x{h}")

    cap.release()
```

```python
video_dir = "/home/projects/sipl-prj10496/project_files/data/source_videos/NL129"  # TODO: update
extensions = ("*.mp4", "*.avi", "*.mov", "*.mkv", "*.MP4")

video_paths = []
for pattern in extensions:
    video_paths.extend(glob.glob(os.path.join(video_dir, "**", pattern), recursive=True))

video_paths = sorted(video_paths)
print(f"Found {len(video_paths)} videos")

frames = []
labels = []

for path in video_paths:
    cap = cv2.VideoCapture(path)
    ok, frame = cap.read()
    cap.release()
    if not ok or frame is None:
        continue
    frames.append(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    labels.append(os.path.basename(path))

cols = 4
rows = max(1, math.ceil(len(frames) / cols))

plt.figure(figsize=(cols * 4, rows * 3))
for i, (frame, label) in enumerate(zip(frames, labels), start=1):
    plt.subplot(rows, cols, i)
    plt.imshow(frame)
    plt.axis("off")
    plt.title(label, fontsize=8)
plt.tight_layout()
plt.show()
```
